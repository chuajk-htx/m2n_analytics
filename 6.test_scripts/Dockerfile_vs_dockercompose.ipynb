# Main Difference between using dockerfile and docker-compose

## dockerfile

This is like a single recipe for you to create an image that you want. For example, a nice laksa dish or kueh lapis.

The benefit of this is to be able to create a single docker image for you to run for your container.

## dockercompose

This is like several of those dishes being put together to coordinate and form a meal, laksa as a main course and kueh lapis for desserts.

docker-compose is for you to run multiple containers easily.
