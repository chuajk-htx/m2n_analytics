## 3. Merging the 2 DataFrames Together

```python
%reset
import pandas as pd

path = (r'C:\Users\TimothyLimChunGao\Documents\GitHub\analytics_platform\4.data_cleanup\4.1.500_500\4.1.1.matching_no_demographics\F-2023_10_10 17_15_16.00000.txt')
with open(path, 'r') as file:
    lines = file.readlines()

# Splitting each line and creating a list of dictionaries
data_list = []
for line in lines:
    values = line.strip().split(',')
    
    for i in range(0, len(values)):
        id1 = values[i][:7]
        id2 = values[i][7:14]
        score = values[i][14:18]

        data_list.append({
            'ID1': id1,
            'ID2': id2,
            'Score': score
        })

# Creating a DataFrame from the list of dictionaries
df1 = pd.DataFrame(data_list)

path = (r'C:\Users\TimothyLimChunGao\Documents\GitHub\analytics_platform\4.data_cleanup\4.1.500_500\4.1.1.matching_no_demographics\demographic_500.txt')
with open(path, 'r') as file:
    lines = file.readlines()

# Splitting each line and creating a list of dictionaries
demo_list = []
for line in lines:
    values = line.strip().split(',')
    
    for i in range(0, len(values)):
        id = values[i][:7]
        id_agegroup = values[i][7:8]
        id_gender = values[i][8:9]
        id_countrycode = values[i][9:10]
        id_race = values[i][10:11]

        demo_list.append({
            'ID': id,
            'ID Age Group': id_agegroup,
            'ID Gender': id_gender,
            'ID Country Code': id_countrycode,
            'ID Race': id_race,
        })
df2 = pd.DataFrame(demo_list)

#print(df1)
#print(df2)

#Merging data for ID1
df_merged = pd.merge(df1, df2, left_on='ID1', right_on='ID', how = 'left')
df_merged.drop(df_merged.columns[[3]], axis=1, inplace=True)
neworder = ['ID1','ID Age Group','ID Gender','ID Country Code','ID Race','ID2','Score']
df_merged=df_merged.reindex(columns=neworder)
df_merged.rename(columns={"ID Age Group": "ID1 Age Group", "ID Gender": "ID1 Gender", "ID Country Code": "ID1 Country Code", "ID Race": "ID1 Race"}, inplace = True)

#print(df_merged)

#Merging data for ID2
df_merged2 = pd.merge(df_merged, df2, left_on='ID2', right_on='ID', how = 'left')

df_merged2.drop(df_merged2.columns[[7]], axis=1, inplace=True)
neworder2 = ['ID1','ID1 Age Group','ID1 Gender','ID1 Country Code','ID1 Race','ID2','ID Age Group','ID Gender','ID Country Code','ID Race','Score']
df_merged2=df_merged2.reindex(columns=neworder2)
df_merged2.rename(columns={"ID Age Group": "ID2 Age Group", "ID Gender": "ID2 Gender", "ID Country Code": "ID2 Country Code", "ID Race": "ID2 Race"}, inplace = True)

df_merged2
```
