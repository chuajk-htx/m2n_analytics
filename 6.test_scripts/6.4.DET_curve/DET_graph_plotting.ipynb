```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import det_curve

# Example genuine and impostor scores
genuine_scores = np.random.normal(0.8, 0.05, 1000)  # Scores for genuine matches
impostor_scores = np.random.normal(0.3, 0.1, 1000)  # Scores for impostor matches

# Labels for genuine and impostor scores
y_true = np.concatenate([np.ones(len(genuine_scores)), np.zeros(len(impostor_scores))])
y_scores = np.concatenate([genuine_scores, impostor_scores])

# Calculate the DET curve
fpr, fnr, thresholds = det_curve(y_true, y_scores)

# Plot the DET curve on a log scale for better visualization
plt.figure(figsize=(8, 8))
plt.plot(fpr, fnr, label="DET Curve", color="blue")
plt.yscale("log")
plt.xscale("log")
plt.xlabel("False Positive Rate (FAR)")
plt.ylabel("False Negative Rate (FRR)")
plt.title("Detection Error Tradeoff (DET) Curve")
plt.grid(True)
plt.legend(loc="best")
plt.show()
```
